```python
st = 1024
a = ['H','T']
ls =[]
for i in a:
    for j in a:
        for k in a:
            ls.append(i+j+k)
ls 
```

```python
st = [1024 for _ in range(8)]
for idx in range(len(ls)): 
    for i in ls[idx]:
        if i == 'H':
            st[idx] *= 2
        else : 
            st[idx] /= 2
st
```
